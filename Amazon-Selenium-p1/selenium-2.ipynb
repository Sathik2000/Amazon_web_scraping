```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
PATH="C:\Program Files\driver\chromedriver.exe"
```

```python
driver=webdriver.Chrome(PATH)
```

```python
driver.get("http://www.amazon.in")
```

```python
search_box=driver.find_element(By.ID,"twotabsearchtextbox")
search_box.clear()
search_box.send_keys('Dell laptops')
driver.find_element(By.ID,"nav-search-submit-button").click()
driver.find_element(By.XPATH,"//span[text()='Dell']").click()

```

```python
lap_names=[]
lap_prices=[]
lap_review=[]
review_count=[]
mrp=[]


# all items
all_products=driver.find_elements(By.XPATH,'//div[@data-component-type="s-search-result"]')

for all_product in all_products:

        # name
        names = all_product.find_elements(By.XPATH, ".//span[@class='a-size-medium a-color-base a-text-normal']")
        for name in names:
            lap_names.append(name.text)

        # price
        prices = all_product.find_elements(By.XPATH, ".//span[@class='a-price-whole']")
        for price in prices:
            lap_prices.append(price.text)
        
        try:
            if len(all_product.find_elements(By.XPATH, ".//span[@class='a-size-base s-underline-text']"))>0:
                # review
                reviews = all_product.find_elements(By.XPATH, ".//span[@class='a-size-base s-underline-text']")
                for review in reviews:
                    lap_review.append(review.text)
                    
            else:
                lap_review.append('0')
        except:
            pass
                
        try:
            if len(all_product.find_elements(By.XPATH, ".//span[@class='a-icon-alt']"))>0:
                # Most review
                review_c=all_product.find_elements(By.XPATH, ".//span[@class='a-icon-alt']")
                for review_cs in review_c:
                    review_count.append(review_cs.text)
                    
            else:
                 review_count.append('0')
                    
        except:
            pass
                

        #MRP
        mrps=all_product.find_elements(By.XPATH, ".//span[@class='a-price a-text-price']")
        for MRP in mrps:
            mrp.append(MRP.text)
    
    

```

```python
print('names',len(lap_names))
print('price',len(lap_prices))
print('review',len(lap_review))
print('review_c',len(review_count))
print('mrp',len(mrp))
print('all_products',len(all_products))
```

```python
import pandas as pd 
data=pd.DataFrame({'Laptop_Names':lap_names,'Laptop_price':lap_prices,'Lap_Total_review_count':lap_review,'MRP':mrp})
```
